```python
import pandas as pd
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
x_train = pd.read_excel('X_train_clf.xlsx')
x_test = pd.read_excel('x_test_clf.xlsx')
y_train = pd.read_excel('y_train_clf.xlsx')
y_test = pd.read_excel('y_test_clf.xlsx')
```

```python
X = np.concatenate((x_train, x_test), axis=0)
Y = np.concatenate((y_train, y_test), axis=0)
```

## **DTC**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Instantiate Decision Tree classifiers
fine_model = DecisionTreeClassifier(criterion='entropy', max_depth=32, max_features=None, random_state=134)
medium_model = DecisionTreeClassifier(criterion='entropy', max_depth=16, max_features=None, random_state=134)
coarse_model = DecisionTreeClassifier(criterion='entropy', max_depth=8, max_features=None, random_state=134)

# Train the fine model
fine_model.fit(x_train, y_train)
y_pred_fine = fine_model.predict(x_test)
accuracy_fine = accuracy_score(y_test, y_pred_fine)
fine_scores = cross_val_score(fine_model, X, Y, cv=10, scoring='accuracy')
fine_accuracy_avg = np.mean(fine_scores)
print("Fine Tree Average Accuracy:", fine_accuracy_avg)
print("Fine Tree Accuracy:", accuracy_fine)

# Train the medium model
medium_model.fit(x_train, y_train)
y_pred_medium = medium_model.predict(x_test)
accuracy_medium = accuracy_score(y_test, y_pred_medium)
medium_scores = cross_val_score(medium_model, X, Y, cv=10, scoring='accuracy')
medium_accuracy_avg = np.mean(medium_scores)
print("Medium Tree Average Accuracy:", medium_accuracy_avg)
print("Medium Tree Accuracy:", accuracy_medium)

# Train the coarse model
coarse_model.fit(x_train, y_train)
y_pred_coarse = coarse_model.predict(x_test)
accuracy_coarse = accuracy_score(y_test, y_pred_coarse)
coarse_scores = cross_val_score(coarse_model, X, Y, cv=10, scoring='accuracy')
coarse_accuracy_avg = np.mean(coarse_scores)
print("Coarse Tree Average Accuracy:", coarse_accuracy_avg)
print("Coarse Tree Accuracy:", accuracy_coarse)
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import numpy as np

# Define k-nearest neighbor models with different weights
fine_knn = KNeighborsClassifier(n_neighbors= 18, weights='uniform')
cubic_knn = KNeighborsClassifier(n_neighbors= 18, weights='uniform',  p=3)
weighted_knn = KNeighborsClassifier(n_neighbors=18, weights='distance')

# Train the fine k-nearest neighbor model
fine_knn.fit(x_train, y_train)
y_pred_fine_knn = fine_knn.predict(x_test)
accuracy_fine_knn = accuracy_score(y_test, y_pred_fine_knn)
fine_knn_scores = cross_val_score(fine_knn, X, Y, cv=10, scoring='accuracy')
fine_knn_accuracy_avg = np.mean(fine_knn_scores)
print("Fine k-NN Average Accuracy:", fine_knn_accuracy_avg)
print("Fine k-NN Accuracy:", accuracy_fine_knn)

# Train the cubic k-nearest neighbor model
cubic_knn.fit(x_train, y_train)
y_pred_cubic_knn = cubic_knn.predict(x_test)
accuracy_cubic_knn = accuracy_score(y_test, y_pred_cubic_knn)
cubic_knn_scores = cross_val_score(cubic_knn, X, Y, cv=10, scoring='accuracy')
cubic_knn_accuracy_avg = np.mean(cubic_knn_scores)
print("Cubic k-NN Average Accuracy:", cubic_knn_accuracy_avg)
print("Cubic k-NN Accuracy:", accuracy_cubic_knn)

# Train the weighted k-nearest neighbor model
weighted_knn.fit(x_train, y_train)
y_pred_weighted_knn = weighted_knn.predict(x_test)
accuracy_weighted_knn = accuracy_score(y_test, y_pred_weighted_knn)
weighted_knn_scores = cross_val_score(weighted_knn, X, Y, cv=10, scoring='accuracy')
weighted_knn_accuracy_avg = np.mean(weighted_knn_scores)
print("Weighted k-NN Average Accuracy:", weighted_knn_accuracy_avg)
print("Weighted k-NN Accuracy:", accuracy_weighted_knn)
```

```python
from sklearn.tree import DecisionTreeClassifier
from imblearn.ensemble import RUSBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Define decision tree base estimator
base_model_fine = DecisionTreeClassifier(criterion='entropy', max_depth=32, max_features=None, random_state=134)
base_model_medium = DecisionTreeClassifier(criterion='entropy', max_depth=16, max_features=None, random_state=134)
base_model_coarse = DecisionTreeClassifier(criterion='entropy', max_depth=8, max_features=None, random_state=134)

# Define RUSBoost classifiers with different configurations
rusboost_fine = RUSBoostClassifier(base_estimator=base_model_fine, n_estimators=100, random_state=42)
rusboost_medium = RUSBoostClassifier(base_estimator=base_model_medium, n_estimators=100, random_state=42)
rusboost_coarse = RUSBoostClassifier(base_estimator=base_model_coarse, n_estimators=100, random_state=42)

# Define Stratified K-Fold
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate fine RUSBoost classifier
accuracy_scores_fine = []
for train_index, test_index in kf.split(X, Y):
    x_train_fold, x_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = Y[train_index], Y[test_index]

    # Train the model
    rusboost_fine.fit(x_train_fold, y_train_fold)

    # Evaluate the model
    y_pred_fold = rusboost_fine.predict(x_test_fold)
    accuracy_fold = accuracy_score(y_test_fold, y_pred_fold)
    accuracy_scores_fine.append(accuracy_fold)

# Calculate average accuracy across folds for fine RUSBoost classifier
accuracy_avg_fine = np.mean(accuracy_scores_fine)
print("Fine RUSBoost Average Accuracy:", accuracy_avg_fine)

# Train and evaluate medium RUSBoost classifier
accuracy_scores_medium = []
for train_index, test_index in kf.split(X, Y):
    x_train_fold, x_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = Y[train_index], Y[test_index]

    # Train the model
    rusboost_medium.fit(x_train_fold, y_train_fold)

    # Evaluate the model
    y_pred_fold = rusboost_medium.predict(x_test_fold)
    accuracy_fold = accuracy_score(y_test_fold, y_pred_fold)
    accuracy_scores_medium.append(accuracy_fold)

# Calculate average accuracy across folds for medium RUSBoost classifier
accuracy_avg_medium = np.mean(accuracy_scores_medium)
print("Medium RUSBoost Average Accuracy:", accuracy_avg_medium)

# Train and evaluate coarse RUSBoost classifier
accuracy_scores_coarse = []
for train_index, test_index in kf.split(X, Y):
    x_train_fold, x_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = Y[train_index], Y[test_index]

    # Train the model
    rusboost_coarse.fit(x_train_fold, y_train_fold)

    # Evaluate the model
    y_pred_fold = rusboost_coarse.predict(x_test_fold)
    accuracy_fold = accuracy_score(y_test_fold, y_pred_fold)
    accuracy_scores_coarse.append(accuracy_fold)

# Calculate average accuracy across folds for coarse RUSBoost classifier
accuracy_avg_coarse = np.mean(accuracy_scores_coarse)
print("Coarse RUSBoost Average Accuracy:", accuracy_avg_coarse)
```

```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define the architecture of the neural network
model_narrow = Sequential([
    Dense(50, input_dim=x_train.shape[1], activation='relu'),
    Dense(50, activation='relu'),
    Dense(50, activation='relu'),
    Dense(50, activation='relu'),
    Dropout(0.1),
    Dense(12, activation='softmax')  # Output layer with softmax for multi-class classification
])

# Compile the model
from keras.optimizers import Adam
optimizer = Adam(lr=0.01)
model_narrow.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Define Stratified K-Fold
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model
accuracy_scores = []
for train_index, test_index in kf.split(X, Y):
    x_train_fold, x_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = Y[train_index], Y[test_index]

    # Convert labels to one-hot encoding
    y_train_fold_encoded = tf.keras.utils.to_categorical(y_train_fold, num_classes=12)
    y_test_fold_encoded = tf.keras.utils.to_categorical(y_test_fold, num_classes=12)

    # Train the model
    model_narrow.fit(x_train_fold, y_train_fold_encoded, epochs=800, batch_size=32, verbose=0)

    # Evaluate the model
    y_pred_narrow_fold = model_narrow.predict(x_test_fold)
    y_pred_narrow_fold_classes = np.argmax(y_pred_narrow_fold, axis=1)
    accuracy_narrow_fold = accuracy_score(y_test_fold, y_pred_narrow_fold_classes)
    accuracy_scores.append(accuracy_narrow_fold)

# Calculate average accuracy across folds
accuracy_avg_narrow = np.mean(accuracy_scores)
print("Average Accuracy:", accuracy_avg_narrow)
```
