```python
# Install RDKit (if not already installed)
!pip install -q rdkit-pypi

# Import modules
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import Image

# SMILES for 5-Azacytidine derivative
smiles = "C1=C(C(=O)NC(=O)N1[C@H]2[C@@H]([C@@H]([C@H](O2)CO)O)O)F"
mol = Chem.MolFromSmiles(smiles)
rdMolDraw2D.Compute2DCoords(mol)

# Optional: Define custom atom colors (e.g., red for oxygen, blue for nitrogen, green for fluorine)
atom_colors = {
    6: (0.1, 0.1, 0.1),  # Carbon - dark gray
    7: (0.0, 0.0, 1.0),  # Nitrogen - blue
    8: (1.0, 0.0, 0.0),  # Oxygen - red
    9: (0.0, 0.5, 0.0),  # Fluorine - green
}

# Create drawer
drawer = rdMolDraw2D.MolDraw2DCairo(600, 600)
options = drawer.drawOptions()

# Font size & bond style
options.atomLabelFontSize = 36
options.bondLineWidth = 3

# Assign custom colors to atom types
for k, v in atom_colors.items():
    options.elemDict[k] = v

# Draw and render
drawer.DrawMolecule(mol)
drawer.FinishDrawing()

# Show image
Image(data=drawer.GetDrawingText())
```
